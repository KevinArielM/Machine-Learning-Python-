```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(30) #fijar semilla
#crear números pseudoaleatorios para los ejes
velocidad_paginas=np.random.normal(3,1,1000)
precio_adq=np.random.normal(50,18,1000)/velocidad_paginas

plt.scatter(velocidad_paginas,precio_adq)
```

```python
x=np.array(velocidad_paginas)
y=np.array(precio_adq)
polinomio=np.poly1d(np.polyfit(x,y,6)) #notar que el grado 1 es la regresión lineal
```

```python
#Ajustar polinomio
xp=np.linspace(0,6,100)
plt.scatter(x,y)
plt.plot(xp,polinomio(xp),c="y")
plt.show()
```

```python
#Para medir si podría ser buen modelo. Notar que a mayo grado obtiene mejor calificación pero se podría sobreajustar
#el modelo y no sería tan buen predictor
from sklearn.metrics import r2_score
r2=r2_score(y,polinomio(x))
print(r2)
```
