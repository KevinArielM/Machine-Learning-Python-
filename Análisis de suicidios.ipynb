Como parte del Día Mundial para la Prevención del Suicidio, hoy 11 de septiembre de 2023, realizo un analisis de datos para entender un poco más este fenómeno. Los datos se recolectaron de World Health Organisation, WHO (OMS, en español). En el dataset se recoge información desde 1950 hasta 2021 en 118 paises por rango de edades (quinquenios).


```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sn, missingno as msno, statistics, matplotlib
```

```python
suicidio_ordenado=suicidios.sort_values(by=["No of Suicides"],ascending=[False])
```

```python
suicidio_ordenado
```

```python
suicidio_anual=suicidios.sort_values(by=["Year"], ascending=[True])
```

```python
suicidio_anual
```

```python
suicidios_1994=suicidios[suicidios['Year'].dt.year == 1994]
```

```python
suicidios_1994
```

```python
suicidios_por_año = suicidios.groupby('Year')['No of Suicides'].sum().reset_index()
```

```python
suicidios_por_año
```

```python
suicidios_por_año_ord=suicidios_por_año.sort_values(by=["Year"], ascending=True)
```

```python
suicidios_por_año_ord
```

```python
plt.figure(figsize=(12, 6))
plt.plot(suicidios_por_año['Year'], suicidios_por_año['No of Suicides'], marker='o', color='skyblue')
plt.xlabel('Año')
plt.ylabel('Número Total de Suicidios')
plt.title('Evolución del Número Total de Suicidios (1950-2020)')
plt.grid(True)

# Muestra el gráfico
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor claridad
plt.tight_layout()
plt.show()
```

Notar que algunos paises no tienen datos en ciertos años. Por ejemplo, China dejó de reportar desde 2018. Es importante para analizar mejor la evolución del número de suicidios.

```python
suicidios_mexico = suicidios.loc[suicidios['Country Name'] == 'Mexico']
```
