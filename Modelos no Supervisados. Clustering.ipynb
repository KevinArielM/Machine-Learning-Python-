```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
datos=pd.read_csv("housecsv.txt")
datos.head()
```

```python
#asignar modelo. Definir clusters
k_media=KMeans(n_clusters=10)
```

```python
etiquetas=k_media.fit_predict(datos[["longitude","latitude"]])
```

```python
plt.scatter(datos.longitude,datos.latitude, c=etiquetas)
```
