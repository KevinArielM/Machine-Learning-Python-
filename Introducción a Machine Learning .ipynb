### Actualización de data frames 

```python
datos.shape #renglones, columnas
```

```python
nuevo=pd.DataFrame({"a":np.arange(1,10),"b":np.arange(21,30)},index=[np.arange(-10,-1)])
nuevo
```

```python
nuevo["c"]=np.arange(91,100) #nueva columna
nuevo
```

```python
nuevo["Suma"]=nuevo.a+nuevo.b+nuevo.c #operaciones entre dataframe
nuevo 
```

```python
nuevo["Resta"]=nuevo.a-nuevo.b-nuevo.c
nuevo
```

```python
nuevo["Producto"]=nuevo.a*nuevo.b*nuevo.c
nuevo
```

```python
nuevo["División"]=nuevo.a/nuevo.b/nuevo.c
nuevo
```

```python
nuevo.shape,nuevo.shape[0],nuevo.shape[1] 
```
