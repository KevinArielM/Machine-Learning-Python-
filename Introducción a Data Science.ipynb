### Escala y agrupamiento de datos

Escala mínimo-máximo $$x´=\frac{x-\min(x)}{\max(x)-\min(x)}$$

```python
datos["house_value_escala"]=(datos.median_house_value-datos.median_house_value.min())/ (datos.median_house_value.max()-datos.median_house_value.min())
```

```python
datos.head()
```

Escala Z-Score $$x´=\frac{x-k}{\sigma}$$



```python
datos["house_value_zscore"]=(datos.median_house_value-datos.median_house_value.mean())/datos.median_house_value.std()
datos.head()
```

Agrupamiento en categorías:

```python
etiquetas=["barato","medio","caro","lujoso"]
datos["Clase"]=pd.cut(datos.house_value_escala,[0,.25,.50,.75,1], 
       labels=etiquetas)
datos.head()
```

```python
#agrupamiento en quartil
datos["Clase_quartil"]=pd.qcut(datos.house_value_escala,4, 
       labels=etiquetas)
datos.head()
```

Escala con paquetería:

```python
from sklearn import preprocessing
salida= preprocessing.MinMaxScaler().fit_transform(datos[["median_house_value"]])
```

```python
#comparar con la escala inicial
np.squeeze(salida) == datos.house_value_escala
```

```python
#hay error de redondeo
print(salida[0], datos.house_value_escala[0])
```

```python
#se concluye que hay precisión numérica
np.allclose(np.squeeze(salida),datos.house_value_escala)
```

```python
#haciendo Z-Score con Scikit Learn
preprocessing.StandardScaler().fit_transform(datos[["median_house_value"]])
```
