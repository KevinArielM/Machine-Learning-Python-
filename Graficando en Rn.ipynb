```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos para el eje X
x = np.linspace(-10, 10, 100)

# Definir la función que deseas graficar
def f(x):
    return x**2

# Calcular los valores de la función para el eje Y
y = f(x)

# Graficar la función
plt.plot(x, y)

# Agregar etiquetas y título al gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de f(x) = x^2')

# Mostrar el gráfico
plt.show()
```

Graficando la densidad exponencial:

```python
# Crear datos para el eje X
x = np.linspace(0, 40, 100)

# Definir la función que deseas graficar
def f(x):
    return 2*np.exp(-2*x)

# Calcular los valores de la función para el eje Y
y = f(x)

# Graficar la función
plt.plot(x, y)

# Agregar etiquetas y título al gráfico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de f(x) = x^2')

# Mostrar el gráfico
plt.show()
```

Y su densidad acumulada

```python
# Definir el parámetro lambda
lmbda = 2


# Crear datos para el eje X
x = np.linspace(0, 5, 10000)  # Rango de valores para la variable aleatoria

# Calcular la distribución acumulada
cdf = 1 - np.exp(-lmbda * x)

# Graficar la distribución acumulada
plt.plot(x, cdf)

# Configurar etiquetas y título del gráfico
plt.xlabel('X')
plt.ylabel('Distribución acumulada')
plt.title('Distribución acumulada de una exponencial con lambda = 2')

# Mostrar el gráfico
plt.show()
```

Para graficar una esfera:

```python
#paquete para graficar en 3d
from mpl_toolkits.mplot3d import Axes3D

# Crear una figura y un subplot en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definir los parámetros de la esfera
center = (0, 0, 0)  # Centro de la esfera
radius = 13  # Radio de la esfera
resolution = 100  # Resolución de la malla para la esfera

# Generar los datos para la esfera. Notar la similitud con la generación de datos en 2d.
u = np.linspace(0, 2 * np.pi, resolution)
v = np.linspace(0, np.pi, resolution)
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Graficar la esfera
ax.plot_surface(x, y, z, color='y', alpha=0.5)

# Configurar etiquetas y título del gráfico
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Esfera')

# Mostrar el gráfico
plt.show()
```

Notar que ahora es posible graficar un gran número de funciones en 2D y en 3D.
